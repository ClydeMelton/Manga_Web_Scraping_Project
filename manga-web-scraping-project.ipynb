```python
# Importing all necassary packages
import numpy as np # Linear algebra
import pandas as pd # Data processing, CSV file I/O (e.g. pd.read_csv)
from bs4 import BeautifulSoup # Web scraping
import requests
import re
```

```python
# Initializing essentail lists
title = [ ]
genre = [ ]
rating = [ ]
manga_page = [ ]
```

```python
# Loop to create an inital list of all available titles
page = 1
while True:
    url = f'https://mangakalot.su/manga?page={page}'
    r = requests.get(url).text
    soup = BeautifulSoup(r, 'html.parser')  
    for j in soup.find_all('div', class_= 'tt'):
        title.append(j.string.replace("\n", ""))
    page += 1
    if page >= 32:
        break
```

```python
# Loop to essentailly duplicate the title list data into the manga_page list
for i in range(len(title)):
    manga_page.append(title[i])
# Loop to convert the original title into a format that can be used as part of the url
for index in range(len(manga_page)): # Loop needs some work to reduce complexity
    manga_page[index] = manga_page[index].replace("‘", '').replace("’", '').replace("'", '')
    manga_page[index] = re.sub('[^a-zA-Z0-9 \n\.]', ' ', manga_page[index])
    manga_page[index] = re.sub('[^a-zA-Z0-9\n\.]', '-', manga_page[index])
    manga_page[index] = manga_page[index].replace('--', '-')
    manga_page[index] = manga_page[index].replace('--', '-')
    manga_page[index] = manga_page[index].strip('-')
    manga_page[index] = manga_page[index].lower()
    index += 1
```

```python
# Loop to read the primary genre and rating for each manga title
for index in range(len(manga_page)):
    url = f'https://mangakalot.su/manga/{manga_page[index]}' # F string holding the current title's page url
    r = requests.get(url).text
    soup = BeautifulSoup(r, 'html.parser')
    for j in soup.find('a', rel= 'tag'): # Loop to obtain the genre information
        genre.append(j.string.replace("\n", ""))
    for j in soup.find('div', class_= 'numscore'): # Loop to obtain the rating information
        rating.append(j.string.replace("\n", ""))
    index += 1
    if len(rating) == len(manga_page): # Ending the loop once the rating list is the same length as the manga_page list
        break
    
```

```python
# Create a dictionary holding the created lists
data = {'title':title, 'genre':genre, 'rating':rating}
# Create a dataframe from the data dictionary
df = pd.DataFrame(data, columns = ['title', 'genre', 'rating'])
```

```python
# Save dataframe as a csv
df.to_csv('manga.csv',index=False)
```
